# Terminating an async calculation with the **TerminateModelChainAsync** endpoint

This notebook demonstrates how to cancel an in-progress asynchronous energy calculation by calling the **`TerminateModelChainAsync`** endpoint using the SolarFarmer Python SDK.

Long-running 3D calculations are submitted via the **`ModelChainAsync`** endpoint and polled for completion. The terminate endpoint allows you to cancel such a calculation at any point — for example when wrong inputs were submitted or the run is no longer needed.

See related SolarFarmer public documentation:
- [Acquiring API key](https://mysoftware.dnv.com/download/public/renewables/solarfarmer/manuals/latest/WebApi/Introduction/ApiKey.html)
- [ModelChainAsync endpoint](https://mysoftware.dnv.com/download/public/renewables/solarfarmer/manuals/latest/WebApi/Endpoints/ModelChainAsyncEndpoint.html)
- [TerminateModelChainAsync endpoint](https://mysoftware.dnv.com/download/public/renewables/solarfarmer/manuals/latest/WebApi/Endpoints/TerminateModelChainAsyncEndpoint.html)

## 1. Terminating an async calculation

When a calculation is submitted via `run_energy_calculation()` with 3D inputs, it runs asynchronously using the `ModelChainAsync` endpoint. Each such run is assigned a unique `instance_id`, which is printed to the console as the calculation starts.

To cancel an in-progress calculation, pass its `instance_id` to `sf.terminate_calculation()`. An optional `reason` can be provided for your own records.

**When to use this:**
- Incorrect inputs were submitted and the run is still in progress
- The calculation is no longer needed and you want to free up cloud resources

> **Note:** Terminating a calculation that has already completed has no effect.


````python
# The instance_id is printed when run_energy_calculation() starts a ModelChainAsync run.
# Replace the value below with the instance_id from your calculation.
instance_id = "ADD INSTANCE ID HERE"

# An optional reason for the termination — recorded for your own reference
reason = "Mistaken inputs"

# Cancel the in-progress async calculation
sf.terminate_calculation(
    instance_id=instance_id,
    reason=reason,
    api_key=api_key,  # This argument can be omitted if `SF_API_KEY` is set in your environment
)
````
